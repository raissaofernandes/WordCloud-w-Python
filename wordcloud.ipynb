# WordCloud com Python

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = open('C:\\Users\\Raissa\\analise\\brutos\\textos_preprocessados.csv', 'r').read() #abrindo o arquivo com os textos
```

```python
#definindo os parâmetros e gerando a nuvem de palavras
wordcloud = WordCloud(max_font_size = 100, max_words = 50, background_color = 'white', width = 1920, height = 1080).generate(text)
```

```python
plt.imshow(wordcloud)
plt.axis("off") #para não mostrar os eixos do gráfico
plt.show()
```
