# WordCloud com Python

```python
from wordcloud import WordCloud
from PIL import Image  
import matplotlib.pyplot as plt
import numpy as np
```

```python
text = open('C:\\Users\\Raissa\\analise\\brutos\\textos_preprocessados.csv', 'r').read() #abrindo o arquivo com os textos
```

### Gerando uma WordCloud básica

```python
#definindo os parâmetros e gerando a nuvem de palavras
wordcloud = WordCloud(max_font_size = 100, max_words = 50, background_color = 'white',
                      width = 1920, height = 1080).generate(text)
```

```python
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off") #para não mostrar os eixos do gráfico
plt.margins(x=0, y=0) 
plt.show()
```

### Gerando uma WordCloud com formato 

```python
pyimg_mask = np.array(Image.open("C:\\Users\\Raissa\\analise\\pyimg.png"))
pyimg_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
transformed_pyimg_mask = np.ndarray((pyimg_mask.shape[0], 
                                     pyimg_mask.shape[1]), np.int32) 

for i in range(len(pyimg_mask)):
    transformed_pyimg_mask[i] = list(map(transform_format, pyimg_mask[i]))
    
transformed_pyimg_mask
```

```python
wCloud = WordCloud(max_font_size = 100, max_words = 100, background_color = 'white',
                   mask = transformed_pyimg_mask)
```

```python
wCloud.generate(text)
```

```python
plt.figure(figsize = [20, 10])
plt.imshow(wCloud, interpolation="bilinear")
plt.axis("off") #para não mostrar os eixos do gráfico
plt.show()
```
